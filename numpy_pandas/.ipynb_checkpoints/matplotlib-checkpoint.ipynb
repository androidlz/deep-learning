## Histogram  柱状图的一种 类别类型的数据

```python
np.random.seed(2019)

mu,sigma = 100,15
x = mu +sigma * np.random.randn(10000)

#bins  每个篮子里放多少东西  patches               blue           透明度
n,bins, patches = plt .hist(x , 5 , density =1 ,facecolor = 'b',alpha = 0.5)



print('n=' ,n)
print('bins= ',bins)
plt.xlabel('Smarts')
plt.ylabel('Probability')
plt.title('Histogram of IQ')
plt.text(60,.025,r'$\mu=100,\ \sigma=15$')
plt.axis([40,160,0,0.03])
plt.grid(True)  #后面有没有网格
plt.show()
```

```python
n, bins, patches = plt.hist(weather_2012_final['Temp (C)'],5,facecolor='g',alpha = 0.5)

plt.xlabel('Temperture in C ')
plt.ylabel('Counts of days')
plt.title('Historram of Temperture in year 2012')
plt.grid(True)
plt.show()
```

# Scatter Plot  点状图

```python
N = 50
x=np.random.rand(N)  #0-1
y=np.random.rand(N)
colors = np.random.rand(N)
area = np.pi*(15*np.random.rand(N)**2)     #0-15 
```

```python
plt.scatter(x,y,s=area,c=colors,alpha=0.5)
```

## Pie Chart  饼状图

```python
labels = 'Cats','Birds','Dogs','Pandas'
size = [25,35,30,10]
explode=(0,0.08,0,0)   #爆炸的意思    图分离出来点距离
```

```python
fig1,ax1 = plt.subplots()
plt.pie(size,explode = explode,labels =labels,autopct = '%1.1f%%')   #autopct  百分比显示的位置
ax1.axis('equal')   #表示出来的是圆形
plt.show()
```
