## Histogram  柱状图的一种 类别类型的数据

```python
np.random.seed(2019)

mu,sigma = 100,15
x = mu +sigma * np.random.randn(10000)

#bins  每个篮子里放多少东西  patches               blue           透明度
n,bins, patches = plt .hist(x , 5 , density =1 ,facecolor = 'b',alpha = 0.5)



print('n=' ,n)
print('bins= ',bins)
plt.xlabel('Smarts')
plt.ylabel('Probability')
plt.title('Histogram of IQ')
plt.text(60,.025,r'$\mu=100,\ \sigma=15$')
plt.axis([40,160,0,0.03])
plt.grid(True)  #后面有没有网格
plt.show()
```
